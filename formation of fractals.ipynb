```python
import matplotlib.pyplot as plt
import math

def draw_fractal(x, y, length, angle, depth):
    if depth == 0:
        return

    x_end = x + length * math.cos(angle)
    y_end = y + length * math.sin(angle)

    plt.plot([x, x_end], [y, y_end], color='black')

    draw_fractal(x_end, y_end, length/2, angle - angle_change, depth-1)
    draw_fractal(x_end, y_end, length/2, angle + angle_change, depth-1)

# Запрашиваем данные у пользователя
length = float(input("Введите длину начального отрезка: "))
angle = float(input("Введите угол поворота (в градусах): "))
depth = int(input("Введите глубину рекурсии: "))

# Преобразуем угол в радианы
angle = math.radians(angle)

# Создаем график и настраиваем его
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')
ax.axis('off')

# Начальные координаты
x_start = 0
y_start = 0

# Угол изменения для каждой ветви
angle_change = math.pi/3

# Рисуем фрактал
draw_fractal(x_start, y_start, length, 0, depth)

# Показываем график
plt.show()
```
